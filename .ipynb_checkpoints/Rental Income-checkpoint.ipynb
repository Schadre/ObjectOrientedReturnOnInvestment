Here we assume that we have a client coming to us asking for an automated Rental Property Calculator. Our client's name is Brandon from a company called "Bigger Pockets". Below, you will find a video of what Brandon usually does to calculate his Rental Property ROI.

Using Visual Studio Code/Jupyter Notebook, and Object Oriented Programming create a program that will calculate the Return on Investment(ROI) for a rental property.

Once completed, commit the project to github and submit the link to this assignment

```python
class ReturnOnInvestment():
    
    def __init__(self):
        self.totalmonthlyincome = 0
        self.totalmonthlyexpenses = 0
        self.totalmonthlycashflow = 0
        self.totalinvestment = 0
        self.returnoninvestment = 0
        
    def income(self):
        self.rent = int(input("Enter Rental Income: "))
        self.laundry = int(input("Enter Laundry Income: "))
        self.storageincome = int(input("Enter Storage Income: "))
        self.miscellaneousincome = int(input("Enter Miscellaneous Income: "))
        
        self.totalmonthlyincome = self.rent + self.laundry + self.storageincome + self.miscellaneousincome
        print(f"The Total Monthly income is {self.totalmonthlyincome}")
        
    def expenses(self):
        response = int(input("Enter Taxes: "))
        response2 = int(input("Enter Insurance: "))
        response3 = int(input("Enter Utilities: "))
        response4 = int(input("Enter HOA: "))
        response5 = int(input("Enter Lawn Service Cost: "))
        response6 = int(input("Enter Repairs: "))
        response7 = int(input("Enter CAPEX: "))
        response8 = int(input("Enter Property management cost: "))
        response9 = int(input("Enter Mortgage: "))
        
        self.totalmonthlyexpenses = int(response) + int(response2) + int(response3) + int(response4) + int(response5) + int(response6) + int(response7) + int(response8) + int(response9)
        print(f"The Total Monthly Expenses are {self.totalmonthlyexpenses}")
        
    def cashFlow(self):
        self.totalmonthlycashflow = self.totalmonthlyincome - self.totalmonthlyexpenses
        print(f"The Total Monthly Cash Flow is {self.totalmonthlycashflow}")
    
    def roi(self):
        response = int(input("Enter Down Payment: "))
        response2 = int(input("Closing Cost: "))
        response3 = int(input("Rehab Budget: "))
        response4 = int(input("Enter Miscellaneous Cost: "))
        
        self.totalinvestment = int(response) + int(response2) + int(response3) + int(response4)
        AnnualCashFlow = self.totalmonthlycashflow * 12
        self.returnOnInvestment = AnnualCashFlow / self.totalinvestment
        print(f"The ROI is {self.returnOnInvestment}")
        
BiggerPockets = ReturnOnInvestment()

while True:
    start = input("Would you like the calculation of ROI ? [Y] Yes, [N] No ")
    if start.upper() == "Y":
        print("Please enter the information accordingly by using numbers only. If you don't, have a budget listed or the information available, input the number 0")
        BiggerPockets.income()
        BiggerPockets.expenses() 
        BiggerPockets.cashFlow()
        BiggerPockets.roi()
    elif start.upper() == "N":
        print("Goodbye.")
        break
    else:
        print("Invalid Input, please try again")
        
```
